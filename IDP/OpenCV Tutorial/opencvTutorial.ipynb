## Rescaling and Resizing the Images and Video frames in OpenCV.
- We usually resize and rescale large Images and video files to prevent computational strain.  
- They contain a lot of information in it and displaying it takes a lot of processing power from the computer.  
- By resizing and rescaling we're trying to get rid of that extra information.  
- **Rescaling a Video:** Modifying its height and width to a particular height & width.  
- Generally it is best practice to downscale our media files as the camera used to capture the media(images and videos) can't go higher that their max capabilities.  

```python
img = cv.imread("testImages/Photos/cat.jpg")
cv.imshow("Cat", img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Rescale a frame
def rescaleFrame(frame, scale=0.75):
    width = int(frame.shape[1] * scale)    
    height = int(frame.shape[0] * scale)
    dimensions = (width, height)
    
    return cv.resize(frame, dimensions, interpolation=cv.INTER_AREA)
    
""" 
shape[1] is the width of any frame
& Shape[0] is the height of any frame.
"""
```

```python
capture = cv.VideoCapture("testImages/Videos/dog.mp4")

while True:
    isTrue, frame = capture.read()

    frame_resized = rescaleFrame(frame, scale=0.2)
    
    cv.imshow("Video", frame)
    cv.imshow("Video Resized", frame_resized)
    
    if cv.waitKey(20) & 0xFF == ord("d"):
        break
        
capture.release()
cv.destroyAllWindows()
```

### Translation  
- Translation means shifting the image along the X-axis and Y-axis.  
- Using translation we can shift an image up, down, left, right.  
- We'll create a translate function.

```python
def translate(img, x, y):
    transMat = np.float32([[1,0,x], [0,1,y]])
    dimensions = (img.shape[1], img.shape[0])
    return cv.warpAffine(img, transMat, dimensions)

""" 
-x -> Left
-y -> Up
 x -> Right
 y -> Down
"""
```

```python
# Right by 100px, Down by 100px
translated = translate(img, 100, 100)
cv.imshow("Translated Image", translated)
cv.imshow("Original", img)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Rotation
- Rotating an image by some angle.
- OpenCv allows us to specify any point (rotation) that we like the image to rotate around.  
- 

```python
def rotate(img, angle, rotPoint=None):
    (height, width) = img.shape[:2]
    
    if rotPoint is None:
        rotPoint= (width//2, height//2)
    
    rotMat = cv.getRotationMatrix2D(rotPoint, angle, 1.0)
    dimensions = (width, height)
    
    
    return cv.warpAffine(img, rotMat, dimensions)
```

```python
rotated = rotate(img, 25)

cv.imshow("Rotated Img", rotated)
cv.imshow("Original", img)
cv.waitKey(0)
cv.destroyAllWindows()
```

## How to Smooth And Blur an image with various Blurring Techniques.
- As discussed before, we generally smooth and blur an image when it has a lot of noise, 
- which can be caused from camera sensors, and are basically problems in lighting when the image the taken.  
  
  
Blurring Methods:
- `GaussianBlur`: This does the same thing as averaging, except that instead of computing the average of all the running pixel intensity, each running pixel is given a particular weight.  
And the average of the products of those weights gives us the value of the true center.  
Using this method, you tend to get less blurring than the `Averaging` method, but it is more natural. 
- But some times GaussianBlur won't suit our purposes.  
- `Averaging`: We define a kernel window over a specific portion of an image, and this window will essentially compute the pixel intensity of the middle pixel of the true center as the average of the surrounding pixel intensities.  
- `MedianBlur`: It is the same thing as averaging except that instead of finding the average we find the median of the surrounding pixels.
Generally it tends to be more effective in reducing noise in an image as compared to averaging.  
- `BilateralBlur`: They are the most effective and are used in a lot of advanced computer vision projects, essentially because how it blurs.  
Traditional Blurring methods basically blur the images without looking at wether we're reducing edges in the image or not.  
This applies the blurring but retains the edges in the image.  


```python
import cv2 as cv
```

```python
from matplotlib.ft2font import MULTIPLE_MASTERS


img = cv.imread("testImages/Photos/cats.jpg")
cv.imshow("Original Img", img)

## Averaging
average = cv.blur( img, (7,7))
cv.imshow("Average Blur", average)

## GaussianBlur
Gauss = cv.GaussianBlur(img, (7,7), 0)
cv.imshow("GaussianBlur", Gauss)


## Median Blur
median = cv.medianBlur(img, 7)
cv.imshow("Median Blur", median)


## Bilateral Blur
bilateral = cv.bilateralFilter(img, 7, 25, 25)
cv.imshow("Bilateral Blur", bilateral)


cv.waitKey(0)
cv.destroyAllWindows()
```

## Bitwise Operators in OpenCV
- There are 4 basic Bitwise operators:
    - AND, OR, XOR & NOT
- They are used in the image processing a lot, e.g. while working with masks.
- At very high level, Bitwise operates operates in a binary manner.  
- i.e. a pixel is turned off if it has a value of zero(0), and turn on if it has value of one(1).  
- 

```python
import cv2 as cv 
import numpy as np 
```

```python
blank = np.zeros((400, 400), dtype="uint8")

rectangle = cv.rectangle(blank.copy(), (30,30), (370, 370), 255, -1)
circle = cv.circle(blank.copy(), (200, 200), 200, 255, -1)


cv.imshow("Rectangle", rectangle)
cv.imshow("Circle", circle)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
## Bitwise And --> intersecting region
bitwise_and = cv.bitwise_and(rectangle, circle)
cv.imshow("Bitwise And", bitwise_and)


# Bitwise OR --> intersecting region and non-intersecting region
bitwise_or = cv.bitwise_or(rectangle, circle)
cv.imshow("Bitwise OR", bitwise_or)


## Bitwise XOR --> non-intersecting region
bitwise_xor = cv.bitwise_xor(rectangle, circle)
cv.imshow("Bitwise XOR", bitwise_xor)




cv.waitKey(0)
cv.destroyAllWindows()
```

## Masking 
- Using the BitWise Operations we can essentially perform masking in OpenCV.
- Masking essentially allows us to focus on certain parts of an image that we like to focus on. 
- 

```python
import cv2 as cv
import numpy as np 

img = cv.imread("testImages/Photos/cats.jpg")
cv.imshow("Original", img)

# Creating a blank image
# Dimensions of the mask has to be of the same size as the original image.  
blank = np.zeros(img.shape[:2], dtype="uint8")
cv.imshow("Blank", blank)


## Creating a mask using the blank image.
mask = cv.circle(blank, (img.shape[1]//2, img.shape[0]//2), 100, 255, -1)
cv.imshow("Mask", mask)

## Creating the masked image
masked = cv.bitwise_and(img, img, mask=mask)
cv.imshow("Masked img", masked)


cv.waitKey(0)
cv.destroyAllWindows()
```

## Threshold an Image and different types of thresholding.
- Thresholding is the binarization of an image. 
- Here, we take an image and convert it to a binary image, i.e. an image where pixels are either 0 (black) or 255(white). 
- Simple Example: 
Take an image and take some particular value that we're going to  call the thresholding value, & compare each pixel of the image to this threshold value. If that pixel intensity is less that the threshold value, we set that pixel intensity to zero(black), and if it is more, then to 255(white).  
- In this sense we can create a binary image just from a regular standalone image.  
- Different types of thresholding: 2 types
    - Simple Thresholding: 
    - Adaptive Thresholding: 
---

#### Simple Thresholding:
- First we'll convert our colored image into Grayscale using `cv.cvtColor` & `cv.COLOR_BGR2GRAY`.  
- We'll use the `cv.threshold()`, which returns 
    - `threshold`: threshold value that we passed 
    - `thresh`: the threshold image

#### The parameters:
- `src:` image src(source)
- `thresh:` the threshold value
- `maxval:` value to set to if the pixel value is grater that thresh value. 
- `type:` type of thresholding(Binary)

---


#### Adaptive Thresholding:
- We get different images when we pass different thresholding values.
- The downside is: we've to manually provide the threshold value. 
- In the advance cases this will not work.  
- We can essentially let the computer find the optimum threshold value by itself,
And using the value that it finds, it binaries the image. 
- 

```python
import cv2 as cv

img = cv.imread("testImages/Photos/cats.jpg")
cv.imshow("Cats", img)


# Simple Thresholding

## Converting the image to Gray scale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# cv.imshow("GrayScale", gray)

threshold, thresh = cv.threshold(gray, 150, 255, cv.THRESH_BINARY)
cv.imshow("Simple Thresholded", thresh)


# Thresholding Inverse
threshold, thresh_inv = cv.threshold(gray, 150, 255, cv.THRESH_BINARY_INV)
cv.imshow("Simple Thresholded Inverse", thresh_inv)



# Adaptive Thresholding.

adaptive_thresh = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 3)
cv.imshow("Adaptive Thresholding", adaptive_thresh)


cv.waitKey(0)
cv.destroyAllWindows()
```
